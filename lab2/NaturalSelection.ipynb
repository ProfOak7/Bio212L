```python
#Import the excel file
df = pd.read_excel (r'./Natural_Selection_Lab.xlsx', index_col=0) 
df
```

```python
fig, ax = plt.subplots()
df1.plot.line(ax = ax, y = 'Straw', marker = 'o')
df1.plot.line(ax = ax, y = 'Chopsticks', marker = 'o')
df1.plot.line(ax = ax, y = 'Forceps', marker = 'o')
df1.plot.line(ax = ax, y = 'Spoon', marker = 'o')
df1.plot.line(ax = ax, y = 'Tongue', marker = 'o')
df1.plot.line(ax = ax, y = 'Fork', marker = 'o')
df1.plot.line(ax = ax, y = 'Knife', marker = 'o')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xlabel("Generation Number")
ax.set_ylabel("Population Size")
ax.set_title("Population Size Over Time")
```
